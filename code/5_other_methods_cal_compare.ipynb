```python
import numpy as np
import pandas as pd
```

```python
data_df = pd.read_excel('./Aggrescan.xlsx')
y_true = data_df['label']
predicted_class = data_df['Aggrescan']

ACC_collecton = []
BACC_collecton = []
Sn_collecton = []
Sp_collecton = []
MCC_collecton = []
precison_collecton = []
F1_collecton = []

y_true = np.array(y_true)
predicted_class = np.array(predicted_class)

from sklearn.metrics import confusion_matrix
import math
# np.ravel() return a flatten 1D array
TP, FP, FN, TN = confusion_matrix(y_true, predicted_class).ravel() # shape [ [True-Positive, False-positive], [False-negative, True-negative] ]
print('TP, FP, FN, TN:', TP, FP, FN, TN)
ACC = (TP+TN)/(TP+TN+FP+FN)
ACC_collecton.append(ACC)
Sn_collecton.append(TP/(TP+FN))
Sp_collecton.append(TN/(TN+FP))
MCC = (TP*TN-FP*FN)/math.pow(((TP+FP)*(TP+FN)*(TN+FP)*(TN+FN)),0.5)
MCC_collecton.append(MCC)
BACC_collecton.append(0.5*TP/(TP+FN)+0.5*TN/(TN+FP))
precison_collecton.append(TP/(TP+FP))
F1 = (2*TP)/(2*TP+FN+FP)
F1_collecton.append(F1)

print('ACC: ', ACC_collecton[0])
print('BACC: ', BACC_collecton[0])
print('Sn: ', Sn_collecton[0])
print('Sp: ', Sp_collecton[0])
print('MCC: ', MCC_collecton[0])
print('Precision: ', precison_collecton[0])
print('F1: ', F1_collecton[0])
```
